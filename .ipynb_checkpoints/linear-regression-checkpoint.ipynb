```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_boston
import matplotlib.pyplot as plt
```

 ## data

```python
boston_dataset = load_boston()
boston = pd.DataFrame(boston_dataset.data, columns=boston_dataset.feature_names)
boston['MEDV'] = boston_dataset.target
```

```python
sns.set(rc={'figure.figsize':(12,8)})
correlation_matrix = boston.corr().round(2)
sns.heatmap(data=correlation_matrix, annot=True, center=0, cmap="coolwarm")
plt.show()
```

 ## things to consider
 * to train a linear regression model, we select features with high positive or negative correlation.
 * when selecting features, avoid having multi-co-linearity within features. See [here](https://stats.stackexchange.com/a/1150) for more.

 ## helper functions

```python
def MSE(yHat, y, return_gradient=True):
    """
    calculate mean squared error
    yHat: 1d list/np.array or float
    y: 1d list/np.array or float
    return_gradient: if or not to return the gradients of the loss
    """
    if isinstance(yHat, list):
        yHat = np.array(yHat)
    if isinstance(y, list):
        y = np.array(y)
    loss = np.sum((yHat - y)**2) / len(y)
    if return_gradient:
        gradient = 2 * (yHat - y) / len(y)
        return loss, gradient
    else:
        return loss
```

 ## LinearRegression

```python
class LinearRegression:
    """
    Linear Regression model
    initialize model parameters
    [n_features]: integer
    """
    def __init__(self, n_features):
        self.weights = np.full((n_features), 0.0)
        self.bias = 0.0

    def train(self, X, Y, iters, learning_rate):
        """
        this is the linear regression part, the training of the model

        X: 2d list/np.array or pd.DataFrame

        Y: 1d list/np.array or pd.Series

        iters: integer

        learning_rate: float
        """
        # type conversion
        if isinstance(X, list):
            X = np.array(X)
        elif isinstance(X, pd.DataFrame):
            X = X.to_numpy()
        
        if isinstance(Y, list):
            Y = np.array(Y)
        elif isinstance(Y, pd.Series):
            Y = Y.to_numpy()

        if len(X.shape) != 2:
            raise TypeError(f'len(X.shape) must be 2 but is {len(X.shape)}')
        if len(Y.shape) != 1:
            raise TypeError(f'len(Y.shape) must be 2 but is {len(Y.shape)}')

        loss_history = []
        for i in range(iters):
            loss, gradient = MSE(self.predict(X), Y)
            for ind in range(len(self.weights)):
                # X[:,ind] is indth column
                # weights gradient has to be multiplied with it's feature because of scalar rule
                grad = np.sum(X[:,ind] * gradient)
                self.weights[ind] -= grad * learning_rate
            # for bias we don't have to multipy by anything because there is no scalar
            self.bias -= np.sum(gradient) * learning_rate
            if i % 10 == 0:
                print(f'iter: {i}, loss: {np.sum(loss)}')
            loss_history.append(np.sum(loss))
        return loss_history

    def predict(self, X):
        """
        return product of X and self.weights + self.bias
        """
        if isinstance(X, list):
            X = np.array(X)
        elif isinstance(X, pd.DataFrame) or isinstance(X, pd.Series):
            X = X.to_numpy()
        elif not isinstance(X, np.ndarray):
            raise TypeError('X must be of type list or np.array')
        if len(X.shape) > 2:
            raise TypeError(f'len(X.shape) must be 1 to 2 but is {len(X.shape)}')
        return np.sum(X * self.weights, axis=len(X.shape)-1) + self.bias

    def reset(self):
        self.weights = np.full((len(self.weights)), 0)
        self.bias = 0
```

```python
model = LinearRegression(2)
losses = model.train(boston[['LSTAT', 'RM']], boston['MEDV'], 200, 0.001)
```

```python
sns.set(rc={'figure.figsize':(8,6)})
sns.lineplot(x = [x for x in range(len(losses))], y = losses)
plt.show()
```

```python
sns.set(rc={'figure.figsize':(14,6)})
plt.subplot(1, 2, 1)
sns.scatterplot(data=boston, x='LSTAT', y='MEDV')
sns.lineplot(x=[x for x in np.arange(boston['LSTAT'].min(), boston['LSTAT'].max(), 0.1)], y=[model.predict([x, boston['RM'].mean()]) for x in np.arange(boston['LSTAT'].min(), boston['LSTAT'].max(), 0.1)])
plt.subplot(1, 2, 2)
sns.scatterplot(data=boston, x='RM', y='MEDV')
sns.lineplot(x=[x for x in np.arange(boston['RM'].min(), boston['RM'].max(), 0.1)], y=[model.predict([boston['LSTAT'].mean(), x]) for x in np.arange(boston['RM'].min(), boston['RM'].max(), 0.1)])
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
fig = plt.figure(figsize=(12, 9))
ax = fig.add_subplot(111, projection='3d')

# to Add a color bar which maps values to colors.
xs = np.arange(boston['LSTAT'].min(), boston['LSTAT'].max(), 0.1)
ys = np.arange(boston['RM'].min(), boston['RM'].max(), 0.1)
xs, ys = np.meshgrid(xs, ys)
zs = np.array([[model.predict([x, y]) for x, y in zip(x_row, y_row)] for x_row, y_row in zip(xs, ys)])
surf=ax.plot_surface(xs, ys, zs, cmap=cm.viridis, linewidth=0.2)

xs = boston['LSTAT']
ys = boston['RM']
zs = boston['MEDV']
ax.scatter(xs, ys, zs, s=50, alpha=0.6, edgecolors='w')

ax.set_xlabel('LSTAT')
ax.set_ylabel('RM')
ax.set_zlabel('MEDV')
plt.show()
```
