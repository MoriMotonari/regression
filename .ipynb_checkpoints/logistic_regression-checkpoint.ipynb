### conclusion

* Pclass / Fare / Cabin are multi-collinear -> only choose one out of them, most collinear one is the Pclass one

* Sex is the most collinear feature

So let's use those two features for the Logistic Regression, both are categorical, but as Pclass is ordered and Sex only binary we can leave it like that. Otherwise we would have to make features for every category value, e.g. a Pclass_is_3 feature#

## helper functions

````python
def sigmoid(x):
    """
    vectorized sigmoid function
    """
    return 1 / (1 + np.exp(-x))

def accuracy(yHat, y):
    assert yHat.shape == y.shape
    return (yHat == y).sum() / yHat.shape[0]

def decide(x):
    res = x >= .5
    res.dtype = np.int8
    return res
````

## Binary Logistic Regression class

````python
class BinaryLogisticRegression:
    """
    Binary Logistic Regression model
    binary means the model has just one output with 1 = 1st category and 0 = 2nd category
    initialize model parameters
    n_features: integer
    """
    def __init__(self, n_features):
        self.weights = np.full((n_features, 1), 0.0)
        self.n_weights = n_features
        self.bias = 0.0

    @staticmethod
    def cross_entropy(yHat, y, return_gradient=True):
        if yHat.shape != y.shape:
            raise ValueError(f'shapes {yHat.shape} and {y.shape} not equal')
        loss = - (y * np.log(yHat) + (1 - y) * np.log(1 - yHat))
        # see https://stats.stackexchange.com/questions/278771/how-is-the-cost-function-from-logistic-regression-derivated for detailed derivation
        if return_gradient:
            gradient = yHat - y
            return loss, gradient
        else:
            return loss


    def train(self, X, Y, iters, learning_rate, info=False):
        """
        this is the logistic regression part, the training of the model

        X: 2d list/np.array or pd.DataFrame

        Y: 1d list/np.array or pd.Series

        iters: integer

        learning_rate: float
        """
        # type conversion
        if isinstance(X, list):
            X = np.array(X)
        elif isinstance(X, pd.DataFrame):
            X = X.to_numpy()
        
        if isinstance(Y, list):
            Y = np.array(Y)
        elif isinstance(Y, pd.Series):
            Y = Y.to_numpy().reshape(len(Y), 1)
        if len(X.shape) != 2 or X.shape[1] != self.n_weights:
            raise TypeError(f'X.shape must be (x, {self.n_weights}) but is {X.shape}')
        if len(Y.shape) != 2 or Y.shape[1] != 1:
            raise TypeError(f'Y.shape must be (x, 1) but is {Y.shape}')

        loss_history = []
        for i in range(iters):
            p = self.predict(X)
            loss, gradient = self.cross_entropy(p, Y)
            for ind in range(self.n_weights):
                # X[:,ind] is indth column
                # partial derivative of weight is it's feature because of scalar rule
                partial_d = X[:, [ind]] # shape (x, 1)
                self.weights[ind, 0] -= (gradient * partial_d * learning_rate).mean()
            # for bias partial derivative is 1
            partial_d = 1
            self.bias -= (gradient * partial_d * learning_rate).mean()
            if info and i % 10 == 0:
                print(f'iter: {i}, loss: {loss.mean()}')
            loss_history.append(loss.mean())
        return loss_history

    def predict(self, X):
        """
        return dot product of X and self.weights + self.bias
        """
        if isinstance(X, list):
            X = np.array(X)
        elif isinstance(X, pd.DataFrame) or isinstance(X, pd.Series):
            X = X.to_numpy()
        elif not isinstance(X, np.ndarray):
            raise TypeError('X must be of type list or np.array')
        if len(X.shape) > 2:
            raise TypeError(f'len(X.shape) must be 1 to 2 but is {len(X.shape)}')
        #return np.sum(X * self.weights.reshape(self.n_weights), axis=len(X.shape)-1) + self.bias
        if len(X.shape) == 1:
            # return as value not array
            return sigmoid((np.dot(X, self.weights) + self.bias).reshape(()))
        return sigmoid(np.dot(X, self.weights) + self.bias) # shape (x, 1)

    def reset(self):
        self.weights = np.full((len(self.weights)), 0.0)
        self.bias = 0.0
````

## evaluation

````python
model = BinaryLogisticRegression(2)
losses = model.train(titanic[['Sex', 'Pclass']], titanic['Survived'], 10000, 0.01)
accuracy(decide(model.predict(titanic[['Sex', 'Pclass']])).reshape(-1), titanic['Survived'])
````

````python
sns.set(rc={'figure.figsize':(8,6)})
sns.lineplot(x = range(len(losses)), y = losses)
plt.show()
````

````python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
fig = plt.figure(figsize=(12, 9))
ax = fig.add_subplot(111, projection='3d')

#0.5 plain
xs = np.arange(titanic['Sex'].min(), titanic['Sex'].max()+1)
ys = np.arange(titanic['Pclass'].min(), titanic['Pclass'].max()+1)
xs, ys = np.meshgrid(xs, ys)
ax.plot_surface(xs, ys, np.array([[0.5 for x, y in zip(x_row, y_row)] for x_row, y_row in zip(xs, ys)]), alpha=0.5, color='red')

#predictions
xs = np.arange(titanic['Sex'].min(), titanic['Sex'].max(), 0.01)
ys = np.arange(titanic['Pclass'].min(), titanic['Pclass'].max(), 0.01)
xs, ys = np.meshgrid(xs, ys)
zs = np.array([[model.predict([x, y]) for x, y in zip(x_row, y_row)] for x_row, y_row in zip(xs, ys)])
ax.plot_surface(xs, ys, zs, cmap=cm.viridis, linewidth=0.2)

#data
xs = titanic['Sex']
ys = titanic['Pclass']
zs = titanic['Survived']
ax.scatter(xs, ys, zs, s=50, alpha=0.6, edgecolors='w')

ax.set_xlabel('Sex')
ax.set_ylabel('Pclass')
ax.set_zlabel('Survived')
ax.view_init(azim=80)
plt.show()
````

## iris

````python
from sklearn.datasets import load_iris
iris_d = load_iris()
print(iris_d['DESCR'])
iris = pd.DataFrame(iris_d['data'], columns=iris_d['feature_names'])
iris['target'] = iris_d['target']
# only use two kinds for binary classification
iris = iris[iris['target'] < 2]
model = BinaryLogisticRegression(len(iris_d['feature_names']))
model.train(iris.drop(columns=['target']), iris['target'], 10000, 0.01)
accuracy(decide(model.predict(iris.drop(columns=['target']))).reshape(-1), iris['target'])
````

````python
fig = plt.figure(figsize=(12, 9))

def build_input(*args):
    res = [iris[f].mean() for f in iris_d['feature_names']]
    for ind, x in args:
        res[ind] = x
    return res

n = len(iris_d['feature_names'])
i = 0
for ind1, f1 in enumerate(iris_d['feature_names']):
    for ind2, f2 in enumerate(iris_d['feature_names']):
        i += 1
        if f1 == f2:
            continue
        ax = fig.add_subplot(n, n, i, projection='3d', visible=f1!=f2)
        if ind1 == 0 or (ind1 == 1 and ind2 == 0):
            ax.set_title(f2)
        #0.5 plain
        xs = [iris[f1].min(), iris[f1].max()]
        ys = [iris[f2].min(), iris[f2].max()]
        xs, ys = np.meshgrid(xs, ys)
        ax.plot_surface(xs, ys, np.array([[0.5 for x, y in zip(x_row, y_row)] for x_row, y_row in zip(xs, ys)]), alpha=0.5, color='red')

        xs = np.arange(iris[f1].min(), iris[f1].max(), 0.05)
        ys = np.arange(iris[f2].min(), iris[f2].max(), 0.05)
        xs, ys = np.meshgrid(xs, ys)
        zs = np.array([[model.predict(build_input((ind1, x), (ind2, y))) for x, y in zip(x_row, y_row)] for x_row, y_row in zip(xs, ys)])
        ax.plot_surface(xs, ys, zs, cmap=cm.viridis, linewidth=0.2)

        xs = iris[f1]
        ys = iris[f2]
        zs = iris['target']
        ax.scatter(xs, ys, zs, s=50, alpha=0.6, edgecolors='w')

        ax.get_xaxis().set_ticks([])
        ax.get_yaxis().set_ticks([])
        ax.zaxis.set_ticks([])
        # ax.view_init(azim=-13, elev=33)
plt.tight_layout()
plt.show()
````
